# N-back model with SRN and LSTM

This notebook aims at understanding similarities/differences between simple recurrent network and LSTM architectures by simulating performance on the N-back task. In N-back, the participant views a series of symbols (e.g. numbers or letters) and must press a button if the currently-viewed symbol matches the one viewed _n_ steps previously. The model should take in a one-hot representation of the currently-viewed symbol and output a single unit that is active when the current item matches the n-back target and inactive otherwise. To do this the model must learn to keep an internal representation of the previously-viewed _n_ items _and_ the order in which they were viewed. I will compare the ability of models using standard simple recurrence vs LSTM recurrence to solve this problem.

## Imports

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.utils import to_categorical #Generates one-hot vectors from single integers
from tensorflow.keras.layers import Input, Dense, LSTM, SimpleRNN #Layer types
from tensorflow.keras.models import Model #Generic model type
```

## Define functions

This function will generate a tensor containing sequences of inputs and targets with the following parameters:
- _n_: The distance at which matches should be judged (ie the n in n-back)
- _numitems_: The total number o symbols in the vocabulary
- _seqlength_: The length of each sequence in training and testing data
- _numpats_: The number of sequences to produce

```python
def make_patterns(n, numitems = 5, seqlength=20, numpats=100):
    x = np.zeros([numpats,seqlength, numitems])
    y = np.zeros([numpats, seqlength,1])
    
    offset = seqlength - n
    
    x = np.random.randint(0,numitems,size=(numpats,seqlength))
    y = np.zeros((numpats,offset))
    m = x[:,0:offset] == x[:,n:seqlength]
    y[m] = 1
    y= np.concatenate([np.zeros((numpats,n)), y], axis=1)
    
    x = to_categorical(x)
    
    
    return x, y
     
        
```

Use function to generate training and testing sets:

```python
n = 3 #Matching distance
nitems = 5 #size of vocabulary
slength = 20 #Length of each sequence
npats = 100 #Number of sequences

[xtrn, ytrn] = make_patterns(n,nitems,slength,npats)
[xtst, ytst] = make_patterns(n,nitems,slength,npats)
xtrn.shape
```

## Build network

The input tensor, excluding the batch dimension (d1), has _slength_ steps in each sequence and _nitems_ possible symbols at each step, so the input shape should be (_slength_, _nitems_).

For the hiden layer we just need to specify the number of hidden units; it will figure out the number of steps in the sequence on its own. Same for the output activation. _Note_ that, to have the hidden layer propagate activation to the output unit on every time-step, we need to set _return_\__sequences_ to **True.** Otherwise the model will take the whole sequences of input and output just a single value at the end.

```python
initem = Input(shape=(slength,nitems)) #Input tensor
hidden = SimpleRNN(units=32, activation='sigmoid', return_sequences=True)(initem)
outact = Dense(1, activation='sigmoid')(hidden)

srn = Model(inputs=initem, outputs=outact)

srn.summary()
```

```python
#Compile model
srn.compile(optimizer='RMSProp', loss='mean_squared_error')
```

```python
H = srn.fit(x=xtrn, y=ytrn, batch_size = 10, epochs = 1000)
```

```python
plt.plot(H.history['loss'])
```

```python
p = srn.predict(xtst)
```

```python
print(p[0,:,0])
```

```python
print(ytst[0,:])
```

```python
plt.scatter(p[0,:,0], ytst[0,:])
```

```python
initem = Input(shape=(slength,nitems)) #Input tensor
hidden = LSTM(units=32, activation='sigmoid', return_sequences=True)(initem)
outact = Dense(1, activation='sigmoid')(hidden)

lstm = Model(inputs=initem, outputs=outact)

lstm.summary()
```

```python
#Compile model
lstm.compile(optimizer='RMSProp', loss='mean_squared_error')
```

```python
H = lstm.fit(x=xtrn, y=ytrn, batch_size = 10, epochs = 2000)
```

```python
plt.plot(H.history['loss'])
```

```python
p = srn.predict(xtst)
plt.scatter(p[0,:,0], ytst[0,:])
plt.scatter(p[1,:,0], ytst[1,:])
```
