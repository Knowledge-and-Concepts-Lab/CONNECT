```python
# Step 1-1: prepare your dataset
import csv
tmp = ""

# Download 771.csv from https://nextml.github.io/caption-contest-data2/summaries/771.csv
with open('771.csv', newline='') as csvfile:
  reader = csv.reader(csvfile)
  next(reader) # skip the header
  for i, row in enumerate(reader):
    txt = row[0].replace('"', '``')
    txt = txt.replace("\n", " ")
    tmp += '{"prompt":"What would be a funny caption that describe this situation?\\n\\n###\\n\\n", "completion":"%s ###"}' % txt
    tmp += "\n"
    if i < 10:
      print(txt)
    if i == 200:
      break
```

```python
# Step 1-2: see how it's formatted
tmp
```

```python
# Step 1-3: save it as a file
text_file = open("completion_list.txt", "w")
text_file.write(tmp)
text_file.close()
```
